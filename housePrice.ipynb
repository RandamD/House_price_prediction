### Split the data into training and testing sets to ensure fair model evaluation

```python
features = ['Square_Footage', 'Num_Bedrooms', 'Num_Bathrooms', 'Garage_Size', 'Neighborhood_Quality','price_Per_Square_Foot','Year_Built']
target = 'House_Price'
x=df[features]
y=df[target]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression # Import LinearRegression instead of LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score

reg = LinearRegression() # Create a LinearRegression object
reg.fit(x_train, y_train) # Fit the model to your training data
```

```python
y_pred = reg.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import r2_score # Correct import statement

r2_score(y_test, y_pred) # Removed extra space before r2_score
```
